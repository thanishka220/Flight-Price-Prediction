```python
import pandas as pd
import numpy as np

def preprocess_data(data):
    # Check if 'Route' exists before handling it
    if 'Route' in data.columns:
        data['Route'] = data['Route'].fillna('Unknown')

    # Check if 'Total_Stops' exists before handling it
    if 'Total_Stops' in data.columns:
        data['Total_Stops'] = data['Total_Stops'].fillna('non-stop')

    # Extract date features
    if 'Date_of_Journey' in data.columns:
        data['Date_of_Journey'] = pd.to_datetime(data['Date_of_Journey'], format='%d/%m/%Y', errors='coerce')
        data['Journey_Day'] = data['Date_of_Journey'].dt.day
        data['Journey_Month'] = data['Date_of_Journey'].dt.month
        data.drop(['Date_of_Journey'], axis=1, inplace=True)

    # Extract time features
    if 'Dep_Time' in data.columns:
        data['Dep_Time'] = pd.to_datetime(data['Dep_Time'], format='%H:%M', errors='coerce')
        data['Dep_Hour'] = data['Dep_Time'].dt.hour
        data['Dep_Minute'] = data['Dep_Time'].dt.minute
        data.drop(['Dep_Time'], axis=1, inplace=True)

    if 'Arrival_Time' in data.columns:
        data['Arrival_Time'] = pd.to_datetime(data['Arrival_Time'].str.split().str[0], format='%H:%M', errors='coerce')
        data['Arrival_Hour'] = data['Arrival_Time'].dt.hour
        data['Arrival_Minute'] = data['Arrival_Time'].dt.minute
        data.drop(['Arrival_Time'], axis=1, inplace=True)

    # Process Duration
    if 'Duration' in data.columns:
        duration_split = data['Duration'].str.extract(r'(?:(\d+)h)?\s*(?:(\d+)m)?')
        data['Duration_Hours'] = duration_split[0].fillna(0).astype(int)
        data['Duration_Minutes'] = duration_split[1].fillna(0).astype(int)
        data.drop(['Duration'], axis=1, inplace=True)

    # Encode Total_Stops
    if 'Total_Stops' in data.columns:
        stop_mapping = {'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4}
        data['Total_Stops'] = data['Total_Stops'].map(stop_mapping).fillna(0).astype(int)

    # Drop redundant columns if they exist
    if 'Route' in data.columns:
        data.drop(['Route'], axis=1, inplace=True)

    # One-hot encoding for categorical variables
    categorical_columns = ['Airline', 'Source', 'Destination', 'Additional_Info']
    existing_categorical = [col for col in categorical_columns if col in data.columns]
    data = pd.get_dummies(data, columns=existing_categorical, drop_first=True)

    return data

# Preprocess the train and test datasets
train_data_processed = preprocess_data(train_data)
test_data_processed = preprocess_data(test_data)

# Save the processed datasets
train_data_processed.to_csv('/content/Data_Train.csv', index=False)
test_data_processed.to_csv('/content/Test_set.csv', index=False)

# Check for missing values
print("\nMissing Values in Training Data:")
print(train_data.isnull().sum())

print("\nMissing Values in Test Data:")
print(test_data.isnull().sum())



```
